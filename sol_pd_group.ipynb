```python
# 4
# With the resulting df from Exercise 1, group it by Product_Name and calculate
# the total Product_Price.
# Define the products DataFrame
products_df = pd.DataFrame({
    'Product_ID': [1, 2, 3],
    'Product_Name': ['iPhone 12', 'Samsung S21', 'Pixel 5'],
    'Product_Price': [999, 899, 699],
})

# Define the sales DataFrame
sales_df = pd.DataFrame({
    'Sale_ID': [1, 2, 3, 4, 5, 6, 7],
    'Sale_Date': pd.to_datetime(['2021-01-01', '2021-02-01', '2021-02-15',
                                 '2020-03-01', '2020-04-01', '2020-02-15',
                                 '2020-01-01']),
    'Customer_ID': [1, 2, 3, 2, 3, 4, 4],
    'Product_ID': [1, 1, 2, 2, 3, 3, 1],
})
# Define the customers DataFrame
customers_df = pd.DataFrame({
    'Customer_ID': [1, 2, 3, 4],
    'Customer_Name': ['John Doe', 'Jane Doe', 'Jim Brown', 'Jake Smith'],
    'Customer_Age': [25, 30, 35, 20],
    'Customer_Email': ['john@doe.com', 'jane@doe.com', 'jim@brown.com', 'jake@smith.com'],
})

mergi = sales_df.merge(products_df, how='inner', on='Product_ID')[['Sale_ID', 'Product_Name', 'Product_Price']]
display(mergi.groupby(['Product_Name', 'Sale_ID']).first())
mergi.groupby(['Product_Name']).aggregate({'Product_Price': ['sum', 'mean']})
```
